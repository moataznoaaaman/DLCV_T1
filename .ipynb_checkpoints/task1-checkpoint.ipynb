```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

### creating the test image

```python
pixels00 = np.ones((256,256)) * 50
pixels01 = np.ones((256,256)) * 100
pixels10 = np.ones((256,256)) * 150
pixels11 = np.ones((256,256)) * 200
tophalf = np.concatenate((pixels00, pixels01), axis=0)
bottomhalf = np.concatenate((pixels10, pixels11), axis=0)
test_image = np.concatenate((tophalf, bottomhalf), axis=1)
test_image = test_image.reshape(512,512)
im=Image.fromarray(test_image).convert("L")
im.save(r"dataset3\testGrayImage.jpg")
```

### adding different noise levels

```python
def add_noise(im,save_path,noise):
    im = im
    im = im + noise
    im = im.astype(int)
    im=Image.fromarray(im).convert("L")
    plt.imshow(im,cmap="gray")
    im.save(save_path)
```

```python
## low level noise
gaussian_noise = np.random.normal(0, 10, size=(512, 512))
add_noise(test_image,r"dataset3\testGrayImage_low.jpg",gaussian_noise)
```

```python
## mid level noise
gaussian_noise = np.random.normal(0, 15, size=(512, 512))
add_noise(test_image,r"dataset3\testGrayImage_mid.jpg",gaussian_noise)
```

```python
## high level noise
gaussian_noise = np.random.normal(0, 20, size=(512, 512))
add_noise(test_image,r"dataset3\testGrayImage_high.jpg",gaussian_noise)
```
