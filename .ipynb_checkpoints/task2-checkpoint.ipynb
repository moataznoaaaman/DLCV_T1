## defining the k-means algorithm (part 1)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
from PIL import Image
from scipy.io import loadmat
np.random.seed(7)
```

```python
y_true = np.zeros((512, 512))
y_true[:256, :256] = 0
y_true[:256, 256:] = 2
y_true[256:, :256] = 1
y_true[256:, 256:] = 3
y_true = y_true.reshape(-1)
```

```python
class k_means:
    def __init__(self, K, data_format, image_path, means = 0):
        self.k = K
        self.data_format = data_format
        self.image_path = image_path
        if means == 0:
            self.means = self.init_means()
        else:
            self.means = means
            
    def init_means(self):
        if(self.data_format == "gray"):
            # mean x its every channel
            means = np.random.uniform(1, 256, size=(self.k,1))
            
        else:
            # mean x its every channel
            means = np.random.uniform(1, 256, size=(self.k,3))
        return means

    def read_data(self):
        # self.incahnce_contrast()
        im = Image.open(self.image_path).convert("L")
        if(self.data_format == "gray"):
            return np.array(im).flatten().reshape(-1,1) # pixels x channels
        else:
            return plt.imread(self.image_path).reshape(-1,3)# pixels x channels
    
    def calculate_distance(self,v1,v2):
        v1 = v1-v2
        v1 = np.square(v1)
        v1 = np.sum(v1)
        v1 = np.sqrt(v1)
        return v1

    def calculate_means(self, points,rnk,means):
        # print(means)
        for i in range(0,means.shape[0]):
            count = 0
            sum = np.zeros(means[i].shape)
            for j in range(0,points.shape[0]):
                if rnk[j] == i:
                    count +=1
                    sum+=points[j]
            # print(count,sum)
            means[i] = sum / count
        return means

    def cluster(self):
        points = self.read_data()
        rnk = np.zeros((points.shape[0],1))
        previous_means = np.array(self.means)
        flag = True
        itrs = 0
        while (flag):
            itrs+=1
            # calculate ranks
            for i in range(0,points.shape[0]):
                rn = -1
                min_dist = np.Inf
                for j in range(0,self.means.shape[0]):
                    # print(points[i],means[j])
                    temp_dist = self.calculate_distance(points[i], self.means[j])
                    if temp_dist < min_dist:
                        rn = j
                        min_dist = temp_dist
                rnk[i] = rn
            # re-calculating the means
            self.means = self.calculate_means(points, rnk, self.means)
            self.means[np.isnan(self.means)] = 0
            if(np.array_equal(self.means,previous_means)):#means.all() == previous_means.all()
                flag = False
                # print(means)
            # print(previous_means)
            print("iterations difference: ",np.absolute(self.means-previous_means))
            previous_means = np.array(self.means)
        return (points,rnk,self.means)

    def get_quad(self, i, j):
        if i <= 255 and j <= 255:
            return 0
        elif i >= 256 and j <= 255:
            return 1
        elif i<= 255 and j >= 256:
            return 2
        else:
            return 3

    def confusion(self, metadata, ground_truth):
        # metadata = self.cluster()
        im = Image.open(ground_truth).convert("L")
        ground_truth = np.array(im)
        confusion_matrix = np.zeros((4,4))
        pixels = metadata[0].reshape(512,512)
        rnks = metadata[1].reshape(512,512)
        for i in range (0,pixels.shape[0]):
            for j in range (0,pixels.shape[1]):
                # pixel = pixels[i]
                rn = np.int16(rnks[i][j])
                mean = metadata[2][rn]
                zero_dist = np.absolute(ground_truth[0][0]-mean)
                one_dist = np.absolute(ground_truth[500][0]-mean)
                two_dist = np.absolute(ground_truth[0][500]-mean)
                three_dist = np.absolute(ground_truth[500][500]-mean)
                dists = np.array([zero_dist, one_dist, two_dist, three_dist])
                quad = self.get_quad(i,j)
                if quad == 0:
                    if zero_dist == np.amin(dists):
                        confusion_matrix[0][0] = confusion_matrix[0][0]+1
                    else:
                        confusion_matrix[0][np.argmin(dists)] = confusion_matrix[0][np.argmin(dists)]+1
                elif quad == 1:
                    if one_dist == np.amin(dists): 
                        confusion_matrix[1][1] = confusion_matrix[1][1]+1
                    else:
                        confusion_matrix[1][np.argmin(dists)] = confusion_matrix[1][np.argmin(dists)]+1
                elif quad == 2:
                    if two_dist == np.amin(dists):
                        confusion_matrix[2][2] = confusion_matrix[2][2]+1
                    else:
                        confusion_matrix[2][np.argmin(dists)] = confusion_matrix[2][np.argmin(dists)]+1
                else:
                    if three_dist == np.amin(dists):
                        confusion_matrix[3][3] = confusion_matrix[3][3]+1
                    else:
                        confusion_matrix[3][np.argmin(dists)] = confusion_matrix[3][np.argmin(dists)]+1
                        
        confusion_matrix = confusion_matrix.astype(int)
        accuracy = self.accuracy(confusion_matrix)
        confusion_df = pd.DataFrame(confusion_matrix, [0, 1, 2, 3], [0, 1, 2, 3])
        confusion_df.index.name = 'Actual'
        confusion_df.columns.name = 'Predicted'
        sns.heatmap(confusion_df,cmap="Blues", annot=True, annot_kws={"size": 20}, fmt='g')
        plt.show()
        print(accuracy)
        return confusion_matrix



    def display(self,pixels,rnk):
        fig = plt.figure(figsize=(10, 7))
        for j in range(0,self.k):
            image = np.array(pixels)
            for i in range(0,image.shape[0]):
                if(rnk[i] != j):
                    image[i] = np.zeros((image[i].shape))
                else:
                    image[i] = self.means[j]
            if(self.data_format == "gray"):
                image = image.reshape(512,512)
            else:
                image = image.reshape(512,512,3)
            fig.add_subplot(1, self.k+1, j+1)
            plt.imshow(image,cmap='gray')#plt.imshow(image,cmap='gray')
            plt.axis('off')
        if(self.data_format == "gray"):
            image = np.array(pixels).reshape(512,512)
        else:
            image = np.array(pixels).reshape(512,512,3)
        
        fig.add_subplot(1, self.k+1, self.k+1)
        plt.imshow(image,cmap='gray')#plt.imshow(image,cmap='gray')
        plt.axis('off')
    def accuracy(self, conf_matrix):
        return "accuracy: " + str((np.diag(conf_matrix).sum()/262144)*100) + "%"
```
