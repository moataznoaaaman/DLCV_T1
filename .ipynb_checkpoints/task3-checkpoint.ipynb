```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
from PIL import Image
import scipy
# from scipy.io import loadmat
# from scipy.stats import norm
np.random.seed(7)
```

```python
class EM:
    def __init__(self, K, mus, sigmas, weights, tol_level, image_path):
        self.k = K # number of clusters
        self.mus = mus # means k x 1
        self.sigmas = np.square(sigmas) # variances K x1
        self. weights = weights # wight of each cluster k x1
        self.tol_level = tol_level # the tolarance level
        self.data = np.array(Image.open(image_path).convert("L")).reshape(-1,1) #data points  n x 1
        self.probabilities = np.array([0])# the probability for each class for each point N X K (2)
        

    def e_step(self): # prob of each data point for each cluster N X K and the new classes weights
        # we need the weights of the classes, parameters of the classes, data point
        
        temp = np.zeros( (self.data.shape[0], self.k)) # N X K
        for i in range(self.k):
            pd = scipy.stats.norm(self.mus[i],self.sigmas[i]).pdf(self.data)
            temp[:,i] = pd.reshape(-1)
        
        numerator = temp * self.weights.reshape(-1) # numerator done N X K (1)
        denominator = numerator.sum(axis=1)[:, np.newaxis] + 1e-8 
        self.probabilities = numerator / denominator # the probability for each class for each point N X K (2)
        self.probabilities[np.isnan(self.probabilities)] = 0
        
        self.weights = self.probabilities.mean(axis=0).reshape(-1,1)
        
    def m_step(self):
        for i in range(self.k):
            weight = self.probabilities[:, [i]]
            total_weight = weight.sum()
            a1 = (self.data * weight).sum(axis=0)
            a2 = total_weight
            if a1 == 0 or a2 == 0:
                self.mus[i] = 0
            else:
                self.mus[i] = a1 / a2
        p1 = np.transpose(np.array([ self.data[:,0] for _ in range(0,self.k)]))
        p2 = np.array([ self.mus.reshape(-1) for _ in range(0,self.data.shape[0])]) # n x k
        p2 = np.square(p1 - p2)
        p2 = p2 * self.probabilities
        p2 = np.sum(p2,axis=0).reshape(-1,1)
        total_weight = np.sum(self.probabilities, axis=0).reshape(-1,1)
        p2 = p2 / total_weight
        p2[np.isnan(p2)] = 0
        self.sigmas = np.sqrt(p2.reshape(-1,1))
        
    def cluster(self):
        stop = False
        self.e_step
        self.m_step
        itrs = 0
        for i in range(0,100000):
            itrs+=1
            old_mus = np.array(np.copy(self.mus))
            old_weights = np.array(self.weights)
            self.e_step()
            self.m_step()
            if np.prod(np.abs(old_mus - self.mus), keepdims=True) < self.tol_level:
                break
        return(self.mus, np.sqrt(self.sigmas), itrs)
        
    def get_quad(self, i, j):
        if i <= 255 and j <= 255:
            return 0
        elif i >= 256 and j <= 255:
            return 1
        elif i<= 255 and j >= 256:
            return 2
        else:
            return 3
            
    def accuracy(self, conf_matrix):
        return "accuracy: " + str((np.diag(conf_matrix).sum()/262144)*100) + "%"

    def classification_matrix(self, means, stds):
        temp = np.zeros( (self.data.shape[0], self.k)) # N X K
        temp1 = np.zeros((self.data.shape[0],1))
        for i in range(self.k):
            prop = scipy.stats.norm(means[i],stds[i]).pdf(self.data)
            temp[:,i] = prop.reshape(-1)
        for i in range(0, temp.shape[0]):
            temp1[i] = np.amax(temp[i])
        temp1 = temp1.reshape(512,512)
        return temp1
        
    def classify_test_image(self, means, stds, ground_truth):
        im = Image.open(ground_truth).convert("L")
        ground_truth = np.array(im)
        confusion_matrix = np.zeros((4,4))
        temp = np.zeros( (self.data.shape[0], self.k)) # N X K
        temp1 = np.zeros((self.data.shape[0],1)).astype(int)
        for i in range(self.k):
            prop = scipy.stats.norm(means[i],stds[i]).pdf(self.data)
            temp[:,i] = prop.reshape(-1)
        for i in range(0, temp.shape[0]):
            temp1[i] = np.argmax(temp[i])
        temp1 = temp1.reshape(512,512)
        # print(temp1.shape)
        for i in range (0,512):
            for j in range (0,512):
                # pixel = pixels[i]
                mean = means[temp1[i][j]]
                zero_dist = np.absolute(ground_truth[0][0]-mean)
                one_dist = np.absolute(ground_truth[500][0]-mean)
                two_dist = np.absolute(ground_truth[0][500]-mean)
                three_dist = np.absolute(ground_truth[500][500]-mean)
                dists = np.array([zero_dist, one_dist, two_dist, three_dist])
                quad = self.get_quad(i,j)
                if quad == 0:
                    if zero_dist == np.amin(dists):
                        confusion_matrix[0][0] = confusion_matrix[0][0]+1
                    else:
                        confusion_matrix[0][np.argmin(dists)] = confusion_matrix[0][np.argmin(dists)]+1
                elif quad == 1:
                    if one_dist == np.amin(dists): 
                        confusion_matrix[1][1] = confusion_matrix[1][1]+1
                    else:
                        f2 = True
                        confusion_matrix[1][np.argmin(dists)] = confusion_matrix[1][np.argmin(dists)]+1
                elif quad == 2:
                    if two_dist == np.amin(dists):
                        confusion_matrix[2][2] = confusion_matrix[2][2]+1
                    else:
                        f3 = True
                        confusion_matrix[2][np.argmin(dists)] = confusion_matrix[2][np.argmin(dists)]+1
                else:
                    if three_dist == np.amin(dists):
                        confusion_matrix[3][3] = confusion_matrix[3][3]+1
                    else:
                        f4 = True
                        confusion_matrix[3][np.argmin(dists)] = confusion_matrix[3][np.argmin(dists)]+1
        confusion_matrix = confusion_matrix.astype(int)
        class1_acc = confusion_matrix[0][0]/np.sum(confusion_matrix[0])
        class2_acc = confusion_matrix[1][1]/np.sum(confusion_matrix[1])
        class3_acc = confusion_matrix[2][2]/np.sum(confusion_matrix[2])
        class4_acc = confusion_matrix[3][3]/np.sum(confusion_matrix[3])
        accuracy = self.accuracy(confusion_matrix)
        confusion_df = pd.DataFrame(confusion_matrix, [0, 1, 2, 3], [0, 1, 2, 3])
        confusion_df.index.name = 'Actual'
        confusion_df.columns.name = 'Predicted'
        sns.heatmap(confusion_df,cmap="Blues", annot=True, annot_kws={"size": 20}, fmt='g')
        plt.show()
        print(accuracy)
        print("class1 accuracy: ", class1_acc * 100,"%")
        print("class2 accuracy: ", class2_acc * 100,"%")
        print("class3 accuracy: ", class3_acc * 100,"%")
        print("class4 accuracy: ", class4_acc * 100,"%")
        return confusion_matrix

        
```

## low level noise

#### e = 0.01

```python
k = 4
phi = np.full(shape=k, fill_value=1/k) 
mu = np.array([[25.],
               [90.],
               [160.],
               [240.]])
sigma = np.array([[1.],
               [1.],
               [1.],
               [1.]])
e = 0.01

em1 = EM(k, mu, sigma, phi, e, r"dataset3\testGrayImage_low.jpg")
results = em1.cluster()
print("means: " , results[0])
print("stds: " , results[1])
print("number of iterations: " , results[2])
classification_matrix = em1.classification_matrix(results[0], results[1])
print("classification matrix: ",classification_matrix)
em1.classify_test_image(results[0],results[1],r"dataset3\testGrayImage.jpg")
```
